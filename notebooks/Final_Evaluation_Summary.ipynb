# Soil Moisture Retrieval - Final Evaluation Summary (V1 vs V2)

This notebook summarizes the evolution of the soil moisture retrieval pipeline from the V1 baseline (SAR/NDVI) to the **V2 Advanced Pipeline (SAR/NDVI/NDWI/LST/Elevation/Rainfall)**.

## 1. V1 Baseline Results (Unified Test Set)
These results were achieved using the standard 6 regional datasets with simple SAR and NDVI features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

v1_results = {
    'Model': ['Random Forest (V1)', 'ANN (V1)', 'Hybrid Ensemble (V1)'],
    'RMSE': [0.0348, 0.0355, 0.0335],
    'R2 Score': [0.875, 0.870, 0.884]
}
display(pd.DataFrame(v1_results))
```

## 2. V2 Advanced Pipeline Results (Peak Performance)
The V2 pipeline adds **LST, NDWI, Elevation, and Rainfall** and uses a **Stacked Ensemble (RF + ANN + Meta-Learner)**.

```python
v2_results = {
    'Model': ['Random Forest (V2)', 'ANN (V2)', 'Stacked Ensemble (V2)'],
    'RMSE': [0.0381, 0.0357, 0.0336],
    'R2 Score': [0.9252, 0.9344, 0.9413]
}
display(pd.DataFrame(v2_results))
```

## 3. Comparison of Ensembles (V1 vs V2)

```python
comp_data = {
    'Phase': ['V1 (Baseline)', 'V2 (Advanced)'],
    'R2 Score': [0.8840, 0.9413]
}
df_comp = pd.DataFrame(comp_data)

plt.figure(figsize=(8, 5))
sns.barplot(data=df_comp, x='Phase', y='R2 Score', palette='viridis')
plt.title("Pipeline Evolution: V1 Hybrid vs V2 Stacked Model R2")
plt.ylim(0.8, 1.0)
for i, val in enumerate(df_comp['R2 Score']):
    plt.text(i, val + 0.005, f"{val:.4f}", ha='center', fontweight='bold')
plt.show()
```
