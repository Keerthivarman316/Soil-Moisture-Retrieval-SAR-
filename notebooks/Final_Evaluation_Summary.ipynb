## 3. Analysis Note on R2 Discrepancies

While the **RMSE remains low (0.03 - 0.07)** across all datasets, some regions like Maharashtra and Dharwad show negative R² scores.

**Why?**
- **Low Target Variance**: In some datasets, the actual soil moisture was nearly a constant value (very low standard deviation).
- **Baseline Comparison**: R² compares the model to a baseline mean prediction. When the data has almost no variance, the baseline is extremely hard to beat mathematically, even if the model's absolute error (RMSE) is small.
